```python
import random
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
quiz_data = pd.DataFrame({
    'question': [
        'What is the capital of France?',
        'What is capital of Andhra Pradesh',
        'What is the chemical symbol for water?',
        'Who developed the theory of relativity?',
        'What is the square root of 81?',
        'What is the capital of India?',
        'Solve: 12 * 12',
        'Who painted the Mona Lisa?',
        'What planet is known as the Red Planet?',
        'What is the boiling point of water?',
        'Http stands for...',
        'AWS stands for...',
        'What is the longest river in the world',
        'What is the name of the Earths natural satellite'
    ],
    'answer': ['Paris', 'Amaravathi', 'H2O', 'Einstein', '9', 'New Delhi', '144', 'Leonardo da Vinci', 'Mars', '100¬∞C','Hypertext Markup Language','Amazon Web Services','Nile','moon'],
    'difficulty': ['easy', 'medium', 'easy', 'medium', 'easy', 'easy', 'medium', 'medium', 'easy', 'medium','Easy','Hard','medium','easy']
})
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(quiz_data['question'])
y = quiz_data['difficulty']

clf = MultinomialNB()
clf.fit(X, y)
```

```python
score = 0
questions_asked = []

print("üß† Welcome to the AI-Powered Quiz Game!\n")

for i in range(5):
    # Random question
    q = quiz_data.sample(1).iloc[0]
    if q['question'] in questions_asked:
        continue
    questions_asked.append(q['question'])

    print(f"\nQ{i+1}: {q['question']}")
    user_ans = input("Your answer: ")

    # Check answer
    if user_ans.strip().lower() == q['answer'].strip().lower():
        print("‚úÖ Correct!")
        score += 1
    else:
        print(f"‚ùå Incorrect. Correct Answer: {q['answer']}")
```

```python
difficulty_pred = clf.predict(vectorizer.transform([q['question']]))[0]
print(f"üìä AI Prediction - This question was: {difficulty_pred.upper()}")

print("\nüéØ Quiz Complete!")
print(f"Your Score: {score}/5")
if score < 3:
    print("üß† Suggestion: Review easier topics to improve.")
else:
    print("üöÄ Great job! Ready for harder questions next time.")
```
